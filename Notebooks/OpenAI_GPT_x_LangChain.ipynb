```python
import pandas as pd

def load_dataset(file_path):
    """
    Load the dataset from a given file path, display a preview, and return the full DataFrame.

    Parameters:
    - file_path (str): The path to the file to be loaded.

    Returns:
    - df (DataFrame): The loaded dataset as a pandas DataFrame.
    """
    try:
        # Loading the dataset
        df = pd.read_csv(file_path)  # Use pd.read_excel or pd.read_json as needed
        
        # For displaying a preview of the dataset
        print("Preview of the loaded dataset:")
        display(df.head())
        
        # The DataFrame 'df' is returned and can be assigned to a variable
        return df
    except Exception as e:
        print(f"An error occurred while loading the dataset: {e}")
        return None

# Usage:
file_path = "/tf/notebooks/Project 2024/Datasets/traffic.csv"
dataset = load_dataset(file_path)

# The'dataset' now holds the full DataFrame loaded from the specified file and can be used with LangChain or other tools.
```

```python
import pandas as pd

# Loading the data
def load_dataset(file_path):
    """
    Load the dataset from a given file path, display a preview, and return the full DataFrame.

    Parameters:
    - file_path (str): The path to the file to be loaded.

    Returns:
    - df (DataFrame): The loaded dataset as a pandas DataFrame.
    """
    try:
        df = pd.read_csv(file_path)  # Use pd.read_excel or pd.read_json as needed
        
        # Displaying a preview of the dataset
        print("Preview of the loaded dataset:")
        display(df.head())
        
        # df is returned and can be assigned to a variable
        return df
    except Exception as e:
        print(f"An error occurred while loading the dataset: {e}")
        return None

# Usage:
file_path = "/tf/notebooks/Project 2024/SUPPLEMENTARY MATERIAL/Datasets/Parking Violations_New York.csv"
df = load_dataset(file_path)
```

```python
from langchain.chat_models import ChatOpenAI
from langchain_experimental.agents import create_pandas_dataframe_agent
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Choosing an OpenAI model
chat = ChatOpenAI(model_name='gpt-4-turbo', temperature=0.7)

# Create the Pandas DataFrame agent
agent = create_pandas_dataframe_agent(chat, df, verbose=True)

# Ask the first question
question = """What's in the data that I uploaded\n"""
print(f"Question 1: {question}")
response_dict = agent.invoke(question)
clean_answer = response_dict['output']
print(f"Answer 1: {clean_answer}")

# Ask the second question
question = """Are the fines paid or not?\n"""
print(f"\nQuestion 2: {question}")
response_dict = agent.invoke(question)
clean_answer = response_dict['output']
print(f"Answer 2: {clean_answer}")

# Ask the third question
question = """Are there any violations over time?\n"""
print(f"\nQuestion 3: {question}")
response_dict = agent.invoke(question)
clean_answer = response_dict['output']
print(f"Answer 3: {clean_answer}\n")


print("\n‚≠ê Visualizations based on the dataset ‚≠ê")
def visualize_data(df):
    """
    Generate visualizations for the loaded dataset.
    """
    if not df.empty:
        numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
        categorical_cols = df.select_dtypes(include=['object', 'category']).columns

        # Visualize distributions for numerical data
        for col in numerical_cols:
            plt.figure(figsize=(10, 4))
            sns.histplot(df[col], kde=True)
            plt.title(f'Distribution of {col}')
            plt.show()

        # Visualize counts for categorical data
        for col in categorical_cols:
            if len(df[col].unique()) <= 10:
                plt.figure(figsize=(10, 4))
                sns.countplot(x=col, data=df, palette='Set3')
                plt.title(f'Count Plot for {col}')
                plt.show()

        # Generate correlation heatmap if there are at least two numerical columns
        if len(numerical_cols) > 1:
            plt.figure(figsize=(10, 8))
            sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='viridis')
            plt.title('Correlation Heatmap')
            plt.show()

# Call visualization after LangChain execution
visualize_data(df)
```
